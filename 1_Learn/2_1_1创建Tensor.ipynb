还可以通过现有的Tensor来创建，此方法会默认重用输入Tensor的一些属性，例如数据类型，除非自定义数据类型

```python
x = x.new_ones(5,3)
print(x)
```

```python
x = torch.randn_like(x,dtype=torch.float32)
print(x)
```

```python
print(x.size())
print(x.shape)
```
